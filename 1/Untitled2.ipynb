проврка по 3 значениям определить является ли они сторонами треуголника

```python
a = int(input())
b = int(input())
c = int(input())

def triangle(a1,a2,a3):
   if a1 < a2 + a3:
        if a2 < a1 + a3:
            if a3 < a1 + a2:
                return 'YES'
            else:
                return 'NO'
        else:
            return 'NO'
   else:
         return 'NO'
triangle(a,b,c)
```

```python
заполнение списка от а до b. если список a>b в порядке возрастания иначе в порядке убывания
```

```python
A = int(input())
B = int(input())
if A <= B:
    for k in range(A, B+1):
        print(k, end=' ')
if A > B:
    for l in range(A, B-1, -1):
        print(l, end=" ") 
```

возведение в степень

```python
def power(a, n):
    res = 1
    for i in range(abs(n)):
        res *= a
    if n >= 0:
        return res
    else:
        return 1 / res

m=float(input())
n=int(input())
print(power(m,n))
```

Сколько чисел бо

```python
a = list(map(int, input().split()))
count = 0
for i in range(1, len(a) - 1):
    if a[i - 1] < a[i] > a[i + 1]:
        count += 1
print(count)
```
